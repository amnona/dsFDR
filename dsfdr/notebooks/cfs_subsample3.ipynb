```python
import warnings   
warnings.simplefilter('ignore') 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib notebook
import sys
sys.path.append('/Users/Serene/calour')
import calour as cl
import numpy as np
import calour.cahelper as cah
import calour.analysis

from scipy.stats import sem
```

```python
cfs=cl.read_taxa('data/cfs.biom','data/cfs.map.txt',sparse=False, filter_orig_reads=1000)
```

```python
cfs.sample_metadata.Subject.value_counts()
```

```python
np.random.seed(10)
cfs_sub1 = cfs.filter_min_abundance(5)
```

```python
cfs_sub1
```

```python
np.random.seed(10)
ds9 = []
bh9 = []
fbh9 = []

for i in range(20):
    cfs_sub9 = cfs_sub1.downsample('Subject', num_keep=38)
    ds = cfs_sub9.diff_abundance('Subject','Control','Patient',fdr_method='dsfdr')
    bh = cfs_sub9.diff_abundance('Subject','Control','Patient',fdr_method='bhfdr')
    fbh = cfs_sub9.diff_abundance('Subject','Control','Patient',fdr_method='filterBH')
    
    if ds is None:
        sig_ds = 0
    else:
        sig_ds = len(ds.feature_metadata.index)
    if bh is None:
        sig_bh = 0
    else:
        sig_bh = len(bh.feature_metadata.index)
    if fbh is None:
        sig_fbh = 0
    else:
        sig_fbh = len(fbh.feature_metadata.index)
        
    ds9.append(sig_ds)
    bh9.append(sig_bh)
    fbh9.append(sig_fbh)
    
DS9 = np.mean(ds9)
BH9 = np.mean(bh9)
FBH9 = np.mean(fbh9)
err_DS9 = sem(ds9)  
err_BH9 = sem(bh9)
err_FBH9 = sem(fbh9)

print(DS9, BH9, FBH9)
print(err_DS9, err_BH9, err_FBH9)
```
